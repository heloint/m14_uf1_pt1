### IMPORTS

```python
import pandas as pd
import numpy as np
```

### DATAFRAME

```python
data: pd.DataFrame = pd.read_csv("../data/cleaned_csv/cases.csv", index_col="case_id")
data.head(15)
```

### FILTERS

- COLUMNAS
  
Muestra las columnas location,species, illnesses

```python
entries_columns = data.loc[:,["location","species", "illnesses"]]
entries_columns.head(15)
```

- FILAS
  
Afectados entre 1998-2000, restaurante, en Florida

```python
entries_files = data.loc[(data["year"] <= 2000) & (data["location"] == "Restaurant") & (data["state"] == "Florida")]
entries_files
```

Mirar el total de afectados en Florida en un restaurante entre 1998-2000 por curiosidad

```python
total_illnesses = entries_files.loc[:,"illnesses"].sum()
print("El total de afectados en Florida entre 1998-2000 y en un restaurante es: ")
total_illnesses
```
